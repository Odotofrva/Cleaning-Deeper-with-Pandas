```python
# What is a Series?
# A Series is a 1D labeled array which holds data
 #dtype: is the metadata (data type)

data = pd.Series([10, 20, 30], index = ['a','b','c'])
print(data)


```

```python
# What is a Dataframe?
# A Dataframe is a 2D tabular labeled array with colunms that can hold different data types.
# Consist of rows and columns 

data = {
    'A': [1,2,3],
    'B': [4,5,6],
    'C': [7,8,9],
    'D': ['x','y','z']
}

df = pd.DataFrame(data)
print(df)
```

```python
# More DataFrame Examples

data = {
    "Name": ["Spiderman", "Superman", "Batman"], 
    "Age": [25, 50, 49]
}

# df = pd.DataFrame(data) # convert to a DataFrame 

df = pd.DataFrame(data, index =["One:", "Two:", "Three:"])  # adding the custom index instead of default numeric.
print(df)
```

```python
# Select a single row using loc

# print(df.loc["One"])


# Select a row using iloc: (This is recommended to use instead of loc because the column names aren't the best conventinal way to look up data)
print(df.iloc[0])   # should choose the index 0 which is Spiderman
print(df.iloc[1])
print(df.iloc[2])
```

```python
# Creating Columns in a DataFrame
df["Skill"] = ["Save Crazy MJ", "Block Cryptonite", "Bingo in the Batcave"]

# Creating a New Row
new_row = pd.DataFrame([{"Name": "Iron Man", "Age": 46, "Skill": "Build Tech"}],
                        index=["Four"])
df = pd.concat([df, new_row])

print(df) # printing new dataframe with column and row added
```
